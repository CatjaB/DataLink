```python
import pandas as pd
#import data
df = pd.read_csv('~/Documents/UEBS/DataLink/covid-19-state-level-data.csv')
#show df to control
df
```

```python
df['cases_per_state'] = df['cases'].groupby(df['state']).transform('sum')
df['deaths_per_state'] = df['deaths'].groupby(df['state']).transform('sum')
```

```python
df.head()
```

```python
del df['date']
del df['fips']
del df['cases']
del df['deaths']
df = df.drop_duplicates()
```
